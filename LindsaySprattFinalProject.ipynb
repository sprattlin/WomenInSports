```python
df.drop(df.iloc[:, 1:202], axis = 1, inplace = True)
df
```

```python
df["Total Male HD Coach FT Coaching Duties"].value_counts()
```

```python
female_ft_total = df["Total Female HD Coach FT Coaching Duties"].sum()     
print(female_ft_total) 
```

```python
female_pt_total = df["Total Female HD Coach PT Coaching Duties"].sum()
print(female_pt_total)
```

```python
male_ft_total = df["Total Male HD Coach FT Coaching Duties"].sum()
print(male_ft_total)
```

```python
male_pt_total = df["Total Male HD Coach PT Coaching Duties"].sum()
print(male_pt_total)
```

```python
sum_head_coaches = female_pt_total + male_ft_total + female_ft_total + male_pt_total
print(sum_head_coaches)
```

# Data Visualization
